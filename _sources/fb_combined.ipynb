# FB Combined

```python
import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt
```

```python
file_path = "facebook_combined.txt"
G = nx.read_edgelist(file_path, nodetype=int)

print("Jumlah Node :", G.number_of_nodes())
print("Jumlah Edge :", G.number_of_edges())
```

```python
partition = community_louvain.best_partition(G)

# Jumlah komunitas
num_communities = len(set(partition.values()))
print("Jumlah Komunitas Terdeteksi:", num_communities)
```

```python
modularity = community_louvain.modularity(partition, G)
print("Modularity:", modularity)
```

```python
from collections import defaultdict

community_groups = defaultdict(list)
for node, comm in partition.items():
    community_groups[comm].append(node)

for cid, members in community_groups.items():
    print(f"Komunitas {cid} | Jumlah anggota: {len(members)}")
```

```python
sample_nodes = list(G.nodes())[:500]
H = G.subgraph(sample_nodes)

pos = nx.spring_layout(H, seed=42)
colors = [partition[n] for n in H.nodes()]

plt.figure(figsize=(10, 8))
nx.draw_networkx(H, pos,
                 node_color=colors,
                 node_size=50,
                 edge_color='gray',
                 linewidths=0.1,
                 with_labels=False)

plt.title("Community Detection with Louvain (Facebook Graph)")
plt.show()
```

```python
for cid, members in community_groups.items():
    print(f"Community ID: {cid} | Member Count: {len(members)}")
    print(f"Node IDs: {members}")
    print("\n")
```

## VisualizeLargerSampleGraph

```python
sample_nodes_larger = list(G.nodes())[:1000]
H_larger = G.subgraph(sample_nodes_larger)

pos_larger = nx.spring_layout(H_larger, seed=42)
colors_larger = [partition[n] for n in H_larger.nodes()]

plt.figure(figsize=(12, 10))
nx.draw_networkx(
    H_larger,
    pos_larger,
    node_color=colors_larger,
    node_size=50,
    edge_color='gray',
    linewidths=0.1,
    with_labels=False
)

plt.title("Community Detection with Louvain (Facebook Graph - Larger Sample)")
plt.show()
```

## SummarizeVisualizedCommunities


```python
community_counts_larger_sample = {}

for node in H_larger.nodes():
    community_id = partition[node]
    community_counts_larger_sample[community_id] = community_counts_larger_sample.get(community_id, 0) + 1

print("Community summary for H_larger subgraph:")
for cid in sorted(community_counts_larger_sample.keys()):
    count = community_counts_larger_sample[cid]
    print(f"Community ID: {cid} | Member Count: {count}")
```
